# 1.Given a list of integers,write a function to return the sum of all prime numbers in that list.

```python
import math
def prime(n):
    if n == 0 or n == 1:
        return False
    else:
        for i in range(2, int(n**0.5)+1):
            if n%i == 0:
                return False
                break
        else:
            return True
def sum_of_primes(l):
    sum_of_primes=0
    for i in l:
        if prime(i):
            sum_of_primes += i
    print("Sum of primes:",sum_of_primes)
l=list(map(int,input().split()))
sum_of_primes(l)
```

# 2.Given a list of integers,write a function to check whether the list is strictly increasing or not.

```python
def increasing(l):
    for i in range(1,len(l)):
        if l[i] < l[i-1]:
            return False
    return True
l=list(map(int,input().split()))
increasing(l)
```

# 3.Write a function to check whether a given list is expanding or not(the difference between adjacent elements should keep on increasing).

```python
def expanding(l):
    d1 = abs(l[1] - l[0])
    d2 = abs(l[2] - l[1])
    
    if d1 >= d2:
        return False
    
    else:
        for i in range(3,len(l)):
            d3 = abs(l[i] - l[i-1])
            if d3 <= d2:
                return False
            d2 = d3
        else:
            return True
l=list(map(int,input().split()))
expanding(l)
```

# 4.write a function to calculate all permutations of a given string.

```python
def toString(lst):
    return ''.join(lst)
def permute(a, l, r):
    if l==r:
        print(toString(a))
    else:
        for i in range(l,r+1):
            a[l], a[i] = a[i], a[l]
            permute(a, l+1, r)
            a[l], a[i] = a[i], a[l] 

if __name__=='__main__':
    string = input("Enter the word : ")
    n = len(string)
    a = list(string)
    print('Permutations of given string  '+string+' are :')
    permute(a, 0, n-1)
```
